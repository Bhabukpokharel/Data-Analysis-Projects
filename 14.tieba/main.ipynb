```python
from PIL import Image
from wordcloud import WordCloud
import jieba
import jieba.analyse
```

```python
jieba.load_userdict("dic.txt")
```

```python
#停用词
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords
stopwords_list = stopwordslist('stop.txt')
```

```python
def get_top_tag(df):
    word_list = ""
    for i in df['content'].unique(): #删除重复帖
        word_list += str(i).replace(" ",'').replace("\n",'')
    for i in df['title']:
        word_list += str(i)
    seg_list = jieba.cut(word_list)
    result = " ".join(seg_list)
    jieba.analyse.set_stop_words('stop.txt') 
    top_tags = jieba.analyse.extract_tags(result, topK=100, withWeight=False, allowPOS=())
    top_tag = ''
    for i in top_tags:
        top_tag += ' ' + i
    return result,top_tag
```

```python
feb_1,feb_2 = get_top_tag(feb)
print(feb_2)
```

```python
jan_1,jan_2 = get_top_tag(jan)
print(jan_2)
```

```python
dec_1,dec_2 = get_top_tag(dec)
print(dec_2)
```

```python
wordcloud = WordCloud(font_path='msyh.ttf',
                            background_color='white',
                          width=600,
                          height=300,
                           max_words = 100,
                            stopwords=stopwords_list,
                         random_state = 30
                         ).generate(jan_2)
plt.figure(figsize = (10,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
