## 数据清洗

```python
#根据实习标题和内容筛选掉非数据分析师类的实习
keywords = ["数据","分析"] #可以调整
real_data = []
for job in data:
    if all(x in data[job]["title"] for x in keywords) or all(x in data[job]["jobdescribe"] for x in keywords):
        #if data[job]["title"].find("数据")>0 and data[job]["jobdescribe"].find("数据")>0:
         #   (data[job]["title"].find("分析")>0 and data[job]["jobdescribe"].find("分析")>0)
        #if keyword in data[job]["title"] or data[job]["jobdescribe"]:
            if data[job] not in real_data:
                real_data.append(data[job])
```

```python
len(real_data) #清洗过后的数值
```

```python
# 调整日薪结构，取平均值，如果写的是面议，就改成None.(只能执行一次，第二次执行会把所有money值清零)
for i in range(len(real_data)):
    try:
        money_range = real_data[i]["daymoney"].split('/')[0].split('-')
        real_data[i]["daymoney"] = (int(money_range[0]) + int(money_range[1])) / 2
    except Exception:
        real_data[i]["daymoney"] = None
```

### 行业分布

```python
import re
```

```python
def split_industry(item):  #同时筛选带有/符号和，符号的数值
    try:
        return re.split('/|,',item)
    except:
        pass
```

```python
data_frame['industry_count'] = data_frame['industry'].apply(split_industry)
```

```python
industry = data_frame['industry_count'].dropna()
```

```python
from collections import Counter
result = Counter({})
for i in industry:
    count = Counter(i)
    if len(count)>1:
        for i in count:
            count[i] = count[i] *1.0 / len(count)
    result = result + count
```

```python
industry_list = result.most_common(15)
```

```python
industry_list
```

```python
industry_name = [industry_list[i][0] for i in reversed(range(15))] #选取前二十的城市
industry_count = [industry_list[i][1] for i in reversed(range(15))]
data = [go.Bar(
            y=industry_name,
            x=industry_count,
            orientation = 'h'
    )]

layout = go.Layout(
    title='哪些行业在招数据分析实习生？',
    xaxis=dict(
        title='实习岗位数量'
    ),
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='industry_basic2-bar')
```

### 岗位城市分布

```python
#因为发现有些岗位是多地，所以不能直接计算计数值，所以要对这部分数据再处理一下
location_dict = data_frame[["location","daymoney"]].to_dict()
```

```python
new_dict = {}
for i in range(len(location_dict["location"])):
    if len(location_dict["location"][i]) > 1:
        cities = location_dict["location"][i].split(',')
        for city in cities:
            if city not in new_dict:
                new_dict[city] = {}
                new_dict[city]["money"] = [location_dict["daymoney"][i]]
                new_dict[city]["count"] = 1./len(location_dict["location"][i]) #重合城市按照1/n计算
            else:
                new_dict[city]["count"] += 1./len(location_dict["location"][i])
                new_dict[city]["money"].append(location_dict["daymoney"][i])
    else:
        if city not in new_dict:
            new_dict[city] = {}
            new_dict[city]["money"] = [location_dict["daymoney"][i]]
            new_dict[city]["count"] = 1
        else:
            new_dict[city]["count"] += 1
            new_dict[city]["money"].append(location_dict["daymoney"][i])
```

```python
a = [(city,new_dict[city]["count"],new_dict[city]["money"]) for city in new_dict]
a.sort(key=lambda x: x[1],reverse=True)
```

```python
city = [a[i][0] for i in range(18)] #选取前二十的城市
count = [a[i][1] for i in range(18)]
```

```python
data = [go.Bar(
            x=city,
            y=count
    )]

layout = go.Layout(
    title='实习岗位所在城市分布',
    xaxis=dict(
        title='城市'
    ),
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='city_basic-bar')
```

```python
traces = []
x_data = [a[i][0] for i in range(10)] #选取前十的城市
y_data = [a[i][2] for i in range(10)]
for xd, yd in zip(x_data, y_data):
    traces.append(go.Box(
            y=yd,
        name = xd
            ))

layout = go.Layout(
    title='不同城市日薪水平',
    xaxis=dict(
        title='城市'
    ),
    yaxis=dict(
        title='日薪/元'
    )
)

fig = go.Figure(data=traces,layout=layout)
py.iplot(fig)
```

## 技能关键词

```python
def keywords(text):
    key = jieba.analyse.extract_tags(text, topK=30, withWeight=False, allowPOS=())
    return key

data_frame['keywords'] = data_frame['jobdescribe'].apply(keywords)
```

```python
def delblank(item): #英文分词
    remove_list = set()
    for i in item:
        if i.find(' ')>0:
            new = i.split(' ')
            for word in new:
                item.append(word)
                remove_list.add(i)  
    for i in remove_list:
        item.remove(i)    
    return item
data_frame['keywords'] = data_frame['keywords'].apply(delblank)
```

```python
text = ''
for word_list in data_frame['keywords']:
    for word in word_list:
        text = text + ',' + word.upper() #英文单词统一大写
```

```python
a = text.encode('gbk').decode('gbk','ignore') #中文编码用GBK
```

```python
with open('stop_word.txt') as f:
    stop_words = f.read().splitlines() #载入stop_word字典，可以自行设置
```

```python
wordcloud = WordCloud(   
                            font_path='msyh.ttf',
                            background_color='white',
                          width=1200,
                          height=600,
                            stopwords=stop_words
                         ).generate(a)



plt.figure(figsize = (20,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
