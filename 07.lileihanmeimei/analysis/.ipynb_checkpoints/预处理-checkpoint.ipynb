```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn import preprocessing
```

```python
df = pd.read_csv("addwatch.csv",index_col=0)
df.head(10)
```

```python
def change_to_num(text):
    if len(text)>2:
        text = text[2:-2]
        return int(text)
    else:
        return None
```

```python
df["已经看过的电影"] = df["已经看过的电影"].apply(change_to_num)
```

```python
df["想看的电影"] = df["想看的电影"].apply(change_to_num)
df.head(10)
```

```python
#把想看的无数据的部分转化成0
def change_nan(s):
    if np.isnan(s):
        s = 0
    return s
df["想看的电影"] = df["想看的电影"].apply(change_nan)
df.head(10)
```

```python
df = df.dropna(axis=0)
df.head(10)
```

```python
def change_time(s):
    return datetime.strptime(s, '%Y-%m-%d').timestamp()
df["注册日期"] = df["注册日期"].apply(change_time)
df.head(10)
```

```python
del df["用户姓名"]
```

```python
df.head(10)
```

```python
df = df.sort_values(by = '注册日期', ascending=False)
df["是否水军"] = ""
df.columns = ['rate','person_page','watched','want','register_date',"isfake"]
df.to_csv("hand.csv",index_label="Index",encoding='utf-8')
```
